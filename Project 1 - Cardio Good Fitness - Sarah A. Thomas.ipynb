# 4.0 - Multivariate Analysis

## 4.1 - Product, Age

__Observations:__

2. TM498 is popular among a similar age range as that of TM195, roughly 19 to 45 (outlier of 47/48 exists).
3. TM798 is popular among a narrower range of customers, roughly 23 to 38 (outliers exist from 40 to 47/48).

## 4.2 - Product, Gender

```python
sns.countplot(x='Product',hue='Gender',data=cardiogood)
plt.show()
```

__Observations:__ Male and female customers nearly equally bought TM195 and TM498 but TM798 was much more popular among male customers.

## 4.3 - Product, Usage

```python
sns.stripplot(cardiogood['Product'],cardiogood['Usage'])
plt.show()
sns.boxplot(cardiogood['Product'], cardiogood['Usage'])
plt.show()
```

__Observations:__ 
1. TM798 customers plan to use their equipment the most at 3-6 times per week.
2. Next, outside a few outliers, TM498 customers plan to use their equipment at 3 times per week.
3. Finally, TM195 customers plan to use their equipment 2-5 times per week.

## 4.4 - Product, Fitness
